```python
import sqlite3
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ana_utils import convert_listlike_cols, db_path, font_size, font_size_large
```

```python
conn = sqlite3.connect(db_path)
df = pd.read_sql("SELECT * FROM results ", conn)
df_datasets = pd.read_sql("SELECT * FROM datasets", conn)
conn.close()
df.head(5)
```

```python
df = convert_listlike_cols(df)
```

```python
df["avg_search_duration"] = df["search_durations"].apply(
    lambda x: sum(x) / len(x) if x else None
)
df["avg_train_samples"] = df["train_samples"].apply(
    lambda x: sum(x) / len(x) if x else None
)
set(df[df["avg_search_duration"]>=2*60*60].sort_values("avg_search_duration")["model"])
```

```python
def facet_plot(df):
    g = sns.FacetGrid(df, col="model", col_wrap=4, sharey=True, sharex=True)
    g.map(sns.lineplot, "avg_train_samples", "avg_search_duration")
    g.set_axis_labels("Avg #Samples", "Avg Duration")
    g.set_titles("{col_name}", size=font_size_large)
    g.set_xlabels(fontsize=font_size)
    g.set_ylabels(fontsize=font_size)
    plt.tight_layout()
    plt.show()
facet_plot(df)
```
